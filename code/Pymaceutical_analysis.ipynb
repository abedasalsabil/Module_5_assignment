### Analysis

- The correlation between mouse weight and average tumor size is 0.84. This represents a positive correlation, which could strongly suggest that when mouse weight increases, the tumour size increases. 
- Based on the barcharts, Capomulin has the most timepoints of most being tested.
- Illustrated from the pie charts, the proportion of male mice is slightly more than female mice.
- The Capomulin treatment line graph suggests that as time increases, the volume of the tumour decreases.

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import linregress

# Study data files
mouse_metadata_path = "Pymaceuticals/data/Mouse_metadata.csv"
study_results_path = "Pymaceuticals/data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset

mouse_data_complete = pd.merge(mouse_metadata, study_results, how="left", on=["Mouse ID", "Mouse ID"])
# Display the data table for preview
mouse_data_complete.head()
```

```python
mice_total = mouse_data_complete["Mouse ID"].nunique()
mice_total
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
mouse_duplicate = mouse_data_complete.loc[mouse_data_complete.duplicated(subset=['Mouse ID', 'Timepoint']),'Mouse ID'].unique()
mouse_duplicate
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
single_dup_mouse = mouse_data_complete.loc[mouse_data_complete["Mouse ID"] == "g989", :]
single_dup_mouse
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_mouse_data = mouse_data_complete[mouse_data_complete['Mouse ID'].isin(mouse_duplicate)==False]
clean_mouse_data
```

```python
# Checking the number of mice in the clean DataFrame.
clean_mouse_num = clean_mouse_data["Mouse ID"].nunique()
clean_mouse_num
```

```python
clean_df = pd.DataFrame(clean_mouse_data)
clean_df
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 

average_tumor = clean_df['Tumor Volume (mm3)'].groupby(clean_df['Drug Regimen']).mean()
median_tumor = clean_df['Tumor Volume (mm3)'].groupby(clean_df['Drug Regimen']).median()
variance_tumor = clean_df['Tumor Volume (mm3)'].groupby(clean_df['Drug Regimen']).var()
std_tumor = clean_df['Tumor Volume (mm3)'].groupby(clean_df['Drug Regimen']).std()
sem_tumor = clean_df['Tumor Volume (mm3)'].groupby(clean_df['Drug Regimen']).sem()

# Assemble the resulting series into a single summary DataFrame.

regimen_df = pd.DataFrame({"Average Tumor Volume": average_tumor, 
                           "Median Tumor Volume": median_tumor, 
                           "Tumor Volume Variance": variance_tumor, 
                           "Tumor Volume Std. Dev.": std_tumor, 
                           "Tumor Volume SEM": sem_tumor})
regimen_df
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, 
# and SEM of the tumor volume for each regimen
# Using the aggregation method, produce the same summary statistics in a single line.

summary_data = clean_df.groupby(['Drug Regimen'])[['Tumor Volume (mm3)']].agg(['mean', 'median', 'var', 'std', 'sem'])
summary_data
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen

scatter_plot = capomulin_df.loc[capomulin_df["Weight (g)"]]
scatter_plot

average_by_mouseID =capomulin_df.groupby(['Mouse ID']).mean()

plt.scatter(average_by_mouseID['Weight (g)'],average_by_mouseID['Tumor Volume (mm3)'], color="blue")
plt.title("Average Tumor Volume vs. Mouse Weight")
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.show
```

```python
# Calculate the correlation coefficient and linear regression model 
correlation = st.pearsonr(average_by_mouseID['Weight (g)'],average_by_mouseID['Tumor Volume (mm3)'])
print(f"The correlation between mouse weight and the average tumor volume is {round(correlation[0],2)}")
```

```python
# for mouse weight and average tumor volume for the Capomulin regimen
x_axis = average_by_mouseID["Weight (g)"]
y_axis = average_by_mouseID["Tumor Volume (mm3)"]

(slope, intercept, rvalue, pvalue, stderr) = linregress(x_axis, y_axis)
regress_values = x_axis * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_axis,y_axis)
plt.plot(x_axis,regress_values,"r-")
plt.annotate(line_eq,(15,30),fontsize=20,color="red")
plt.title("Weight vs Average Tumor Volume for Capomulin")
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3))')
plt.show()                           
```
